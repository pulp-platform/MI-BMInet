```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dr_str = f'dr_01'
lr_str = f'lr_1e-4'    

#Names of directories to access
directory = f'5_fold_cv_whole_dataset/'

print(f'LearningRate: 1e-4, Dropout: 0.1')
#Names of files to access
training_loss_str   = f'{directory}train_loss_{lr_str}_{dr_str}_avg.csv'
validation_loss_str = f'{directory}valid_loss_{lr_str}_{dr_str}_avg.csv'
training_accu_str    = f'{directory}train_accu_{lr_str}_{dr_str}_avg.csv'
validation_accu_str  = f'{directory}valid_accu_{lr_str}_{dr_str}_avg.csv'
'''
#Read Data
training_loss_df    = pd.read_csv(training_loss_str)
validation_loss_df  = pd.read_csv(validation_loss_str)
training_acc_df     = pd.read_csv(training_acc_str)
validation_acc_df   = pd.read_csv(validation_acc_str)
'''
#Read Data
training_loss_df    = np.loadtxt(training_loss_str)
validation_loss_df  = np.loadtxt(validation_loss_str)
training_accu_df     = np.loadtxt(training_accu_str)
validation_accu_df   = np.loadtxt(validation_accu_str)
###########Plot, label, save###########
# LOSS
plt.plot(training_loss_df,label='Training')
plt.plot(validation_loss_df,label='Validation')
plt.title(f'Loss: LR={10**(-4)}, DR=0.1')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig(f'{directory}/5fold_avg_loss.pdf')
plt.show()
# ACCURACY
plt.plot(training_accu_df,label='Training')
plt.plot(validation_accu_df,label='Validation')
plt.title(f'Accuracy: LR={10**(-4)}, DR=0.1')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig(f'{directory}/5fold_avg_accu.pdf')
plt.show()
```
